```python
%pylab inline
def rk4(y, time , dt ,derivs):
    '''
        This function moves the value of 'y' forward by a single
        step of size 'dt', using a forth-order Runge-Kutta algorithm. 
        This particular algorithm is equivalent to find the average of the slope at time t and 
        at time (t+dt), and using the average slope to find the new value of y
    '''
    k1 = dt * derivs(y,time)
    k2 = dt * derivs(y+0.5*k1,time + 0.5*dt)
    k3 = dt * derivs(y+0.5*k2,time + 0.5*dt)
    k4 = dt * derivs(y+k3,time + dt)
    y_next = y + 1./6 * (k1 + 2*k2 + 2*k3 +k4)
    return y_next

def pendulum (state, time):
    g0 = state[1]
    g1 = -gravity/length*np.sin(state[0]) 
    return np.array([g0,g1])

N = 100000
x0 = np.pi/6
v0 = 0
tau = 15.0
dt = tau/float(N-1)
beta = 1.0

length = 1.0
gravity = 9.8
#————————————————————————
#分配内存
time = np.linspace(0, tau, N)
y = np.zeros([N,2])#先给y分配好空间！因为一共跑N次，自然是N*2的矩阵

y[0,0] = x0
y[0,1] = v0
#——————————————————————
#循环求解
for i in range(N-1):
    y[i+1] = rk4(y[i],time[i],dt,pendulum)
xdata = [y[i,0] for i in range(N)]
vdata = [y[i,1] for i in range(N)]

plt.plot(time, xdata,"r:",label="x")
#plt.plot(time, vdata,label = "v")
plt.xlabel("time")
plt.ylabel("position, velocity")
plt.legend(loc=1)
plt.grid()
```

```python
from pylab import *
from scipy.integrate import odeint

steps = 1000
theta_initial = pi/6.0
v_initial = 0.0
#omega = 1.0
beta = 0.5
amp = 1.2
omega_d = 0.62
time_step = 2.0*pi/(omega_d*steps)

N = 300
skip = 100

initial_state = array([theta_initial, v_initial])
time = arange(0.0,N*(2.0*pi)/omega_d, time_step)

def DE_pendulum(y,time):
    
    g0 = y[1]
    g1 = -sin(y[0]) - beta * y[1] +amp*sin(omega_d*time)
    return array([g0,g1])

answer = odeint(DE_pendulum, initial_state, time)
answer = answer[skip*steps:]

for i ,position in enumerate(answer[:,0]):
    while position >pi:
        position = position - 2.0*pi
    while position <-pi:
        position = position + 2.0*pi
        
    answer[i,0] = position
    
offset = 50
max_index = (N-skip)*steps - offset

P_thetas = []
P_omegas = []

for j in range(offset, max_index, steps):
    P_thetas.append([answer[j,0]])
    P_omegas.append([answer[j,1]])
    
plot(answer[:,0],answer[:,1],"c:")
plot(P_thetas, P_omegas, "b:")
xlabel("angle")
ylabel("angular velocity")
xlim(-pi,pi)
show()
    
```
