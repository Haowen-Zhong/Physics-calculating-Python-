# ODE

## Euler Method
$$x(x+\Delta t)=x(t)+\dot{x}\Delta t+(\ddot{x}\frac{\Delta t^2}{2})$$
$$Error \sim \mathcal{O}(\Delta t^2)$$
E.g.1:
$$\ddot{x} = -\omega^2x-\beta\dot{x}$$
Let $$v\equiv \dot{x}$$ 
$$\dot{v}=-\omega^2x-\beta v$$
这两个一阶的微分方程联立就等价一个二阶ODE

真刀真枪地解一个:
$$\ddot{x} = -g$$
$$\dot{x}=v$$
$$\dot{v}=-g$$

$$\Rightarrow x_{i+1} = x_i+\dot{x}\Delta t=x_i+v\Delta t$$
$$\Rightarrow v_{i+1} = v_i+\dot{v}\Delta t=v_i-g\Delta t$$
这两个方程具有非常对称的形式因此可以写成：

# $$y_{i+1}(t) = y_i(t) + \dot{y}(t)\Delta t$$ 
$$y=\begin{bmatrix} x\\v \end{bmatrix}$$

$$\dot{y}=\begin{bmatrix} v\\-g \end{bmatrix}$$

# Runge-Kutta Methods
## --The most popular and stable general technique of solving ODE's is a set of methods known as Runge-Kutta methods

Define $y$ as the function we're looking for, and $g$ as the derivative of $y$. Now $g$ is in general a function of both $y$ and $t$, so the second derivative of $y$ is, use the chain rule:

$$\ddot{y} = \frac{d}{dt}\dot{y}$$
$$\equiv\frac{d}{dt}[g(y,t)]$$
$$=\frac{\partial g}{\partial t} + \frac{\partial g}{\partial y}\frac{dy}{dt}$$
$$\boxed{\Rightarrow\ddot{y}=g_t + g_yg}$$
同样的道理我们就可以计算出来$$\boxed{\dddot{y} = g_{tt} +2gg_{ty} + g^2g_{yy}+gg_y^2+g_tg_y}$$
故技重施，采用$Taylor\  expansion$ of $y(t+\Delta t)$ is then:
$$\boxed{y(t+\Delta t) = y(t)
                  +g\Delta t   
                  +\frac{(\Delta t)^2}{2!}(g_t+g_y g)
                  +\frac{(\Delta t)^3}{3!}(g_{tt} +2gg_{ty} + g^2g_{yy}+gg_y^2+g_tg_y)
                  +\mathcal{O}(\Delta t)^4\ \ }$$
我们同样也可以用另一组多项式来展开:
$$y(t+\Delta t) = y(t) + \alpha_1k_1+\alpha_2k_2+...+\alpha_nk_n$$
$$k_1 = \Delta tg(y,t)$$
$$k_2 = \Delta tg(y+\nu_{21}k_1,t+\nu_{21}\Delta t)$$
$$k_1 = \Delta tg(y+\nu_{31}k_1+\nu_{32}k_2,t+\nu_{31}\Delta t+\nu_{32}\Delta t)$$
$$.\\
.\\
.$$
$$k_n = \Delta tg(y+\sum_{\ell=1}^{n-1}\nu_{n\ell k_\ell},t+\sum_{\ell=1}^{n-1}\nu_{n\ell})$$
我们需要确定出$alpha,\nu$这些参数，怎么确定？利用两种展开方式的对比
只展开到两阶：
$$y(t+\Delta t) = y + \alpha_1k_1+\alpha_2k_2$$
$$=y+\alpha_1[\Delta tg(y,t)]+\alpha_2[\Delta tg(y+\nu_{21}k_1,t+\nu_{21}\Delta t)]$$
第一项和第二项已经OK了，因此只需要把第三项展开来：
$$k_2 = \Delta tg(y+\nu_{21}k_1,t+\nu_{21}\Delta t)$$
$$=\Delta tg[g + \nu_{21}k_1g_y+\nu_{21}\Delta tg_t+\mathcal{O}(\Delta t^2)]$$
$$=\Delta t g + \nu_{21}\Delta t^2gg_y+\nu_{21}\Delta t^2g_t+\mathcal{O}(\Delta t^3)$$
$$\boxed{\Rightarrow y(t+\Delta t)=y+[(\alpha_1+\alpha_2)g]\Delta t+[\alpha_2\nu_{21}(gg_y+g_t)]\Delta t^2\ \ }$$
$$\Rightarrow \alpha_1+\alpha_2=1, \alpha_2\nu_{21}=\frac{1}{2}$$
三个未知数，两个方程这是一个不定方程，我们不妨就取$\nu_{21}=1,\alpha_i=\frac{1}{2}$
于是就得到了二阶龙格库塔法：
$$\boxed{y(t+\Delta t) = y + \frac{1}{2}k_1+\frac{1}{2}k_2+\mathcal{O}(\Delta t^3)}$$
$$\boxed{k_1 = \Delta t g(y,t)}$$
$$\boxed{k_2 = \Delta t g(y+k_1,t+\Delta t)}$$

```python
def rk2(y, time , dt ,derivs):
    '''
        This function moves the value of 'y' forward by a single
        step of size 'dt', using a second-order Runge-Kutta algorithm. 
        This particular algorithm is equivalent to find the average of the slope at time t and 
        at time (t+dt), and using the average slope to find the new value of y
    '''
    k0 = dt * derivs(y,time)
    k1 = dt * derivs(y+k0,time + dt)
    y_next = y + 0.5 * (k0 + k1)
    
    return y_next
def SHO(y,t):
    dot0 = y[1]
    dot1 = -k/m*y[0] - gravity
    return np.array([dot0,dot1])


N = 1000
x0 = 0.0
v0 = 0.0
k = 3.5
m = 0.2
gravity = 9.8

tau = 50
dt = tau/float(N-1)

state = np.zeros([N,2])
time = np.linspace(0,tau,N)

state[0,0] = x0
state[0,1] = v0

for i in range(N-1):
    state[i+1] = rk2(state[i], time[i] , dt, SHO)
x = [state[i,0] for i in range(N)]
v = [state[i,1] for i in range(N)]
    
plt.plot(time, x,"r:",label="x")
#plt.plot(time, v,label = "v")
plt.xlabel("time")
plt.ylabel("position, velocity")
plt.legend(loc=1)
plt.grid()    
```

# 4-order Runge-Kutta Method:
$$y(t+\Delta t)= y(t) + \frac{1}{6}(k_1+2k_2+2k_3+k_4)+\mathcal{O}(\Delta t ^4)$$
$$k_1 = g(y,t)\Delta t$$
$$k_2 = g(y+\frac{1}{2}k_1,t+\frac{1}{2}\Delta t)\Delta t$$
$$k_3 = g(y+\frac{1}{2}k_2,t+\frac{1}{2}\Delta t)\Delta t$$
$$k_2 = g(y+k_3,t+\Delta t)\Delta t$$


```python
def rk4(y, time , dt ,derivs):
    '''
        This function moves the value of 'y' forward by a single
        step of size 'dt', using a forth-order Runge-Kutta algorithm. 
        This particular algorithm is equivalent to find the average of the slope at time t and 
        at time (t+dt), and using the average slope to find the new value of y
    '''
    k1 = dt * derivs(y,time)
    k2 = dt * derivs(y+0.5*k1,time + 0.5*dt)
    k3 = dt * derivs(y+0.5*k2,time + 0.5*dt)
    k4 = dt * derivs(y+k3,time + dt)
    y_next = y + 1./6 * (k1 + 2*k2 + 2*k3 +k4)
    return y_next

def SHO(y,t):
    dot0 = y[1]
    dot1 = -k/m*y[0] - gravity
    return np.array([dot0,dot1])

N = 1000
x0 = 0.0
v0 = 0.0
k = 3.5
m = 0.2
gravity = 9.8

tau = 10
dt = tau/float(N-1)

state = np.zeros([N,2])
time = np.linspace(0,tau,N)

state[0,0] = x0
state[0,1] = v0

for i in range(N-1):
    state[i+1] = rk4(state[i], time[i] , dt, SHO)
x = [state[i,0] for i in range(N)]
v = [state[i,1] for i in range(N)]
    
plt.plot(time, x,"r:",label="x")
#plt.plot(time, v,label = "v")
plt.xlabel("time")
plt.ylabel("position, velocity")
plt.legend(loc=1)
plt.grid()    
```

```python
#Eg2 这个例子是一个巧合，因为加速度和时间本身没有关系，只和x有关系
def rk4(y, time , dt ,derivs):
    '''
        This function moves the value of 'y' forward by a single
        step of size 'dt', using a forth-order Runge-Kutta algorithm. 
        This particular algorithm is equivalent to find the average of the slope at time t and 
        at time (t+dt), and using the average slope to find the new value of y
    '''
    k1 = dt * derivs(y,time)
    k2 = dt * derivs(y+0.5*k1,time + 0.5*dt)
    k3 = dt * derivs(y+0.5*k2,time + 0.5*dt)
    k4 = dt * derivs(y+k3,time + dt)
    y_next = y + 1./6 * (k1 + 2*k2 + 2*k3 +k4)
    return y_next

def SpringMass(state, time):
    g0 = state[1]
    
    if g0>0:
        g1 = -k/m * state[0] - gravity * m 
    else :
        g1 = -k/m * state[0] + gravity * m 
    return np.array([g0,g1])
N = 500
x0 = 0.2
v0 = 0.0

tau = 3.0

dt = tau/float(N-1)

k = 42.0
m = 0.25
gravity = 9.8
mu = 0.15

y = np.zeros([N,2])

y[0,0] = x0
y[0,1] = v0

for i in range(N-1):
    y[i+1] = rk4(y[i] , 0, dt, SpringMass)
time = np.linspace(0, tau, N)
plt.plot(time, y[:,0], 'b-' , label="postion")
plt.xlabel("time")
plt.ylabel("postion")
```
