```python
%pylab inline
from sklearn import datasets
boston = datasets.load_boston()
print(boston.DESCR)
```

```python
Y = boston.target
X = boston.data
```

```python
import math
half = math.floor(len(Y)/2)
train_X = X[:half]
train_Y = Y[:half]
test_X = X[half:]
test_Y = Y[half:]
```

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()
```

```python
linear_model.LinearRegression?
```

```python
#Example
import numpy as np
from sklearn.linear_model import LinearRegression
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])


# y = 1 * x_0 + 2 * x_1 + 3..interesting
y = np.dot(X, np.array([1, 2])) + 3


reg = LinearRegression().fit(X, y)
reg.score(X, y)


reg.intercept_ # doctest: +ELLIPSIS
reg.predict(np.array([[3, 5]]))

```

```python
clf.fit(train_X,train_Y)
```

```python
Y_lr_pred = clf.predict(test_X)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test_Y,Y_lr_pred) 
print(mse)
```

```python
f, ax = plt.subplots()
ax.plot(test_Y,Y_lr_pred - test_Y,'o',alpha=0.5)
ax.set_title("Linear Regression Residuals - MSE = %.1f" % mse)
ax.set_xlabel("True Median House Price ($1,000)")
ax.set_ylabel("Residual")
ax.hlines(0,min(test_Y),max(test_Y),color="red")
```

## *k*-Nearest Neighbor (KNN) Regression

It's easy to understande the criteria to regression of this algorithm

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
X = np.array([[-3, -1], [-4, -5],[-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
nbrs = NearestNeighbors(n_neighbors=3, algorithm='ball_tree').fit(X)
distances, indices = nbrs.kneighbors(X)
```

```python
distances
```

```python
plt.plot(X[:,0],X[:,1],"rx")
```

```python
nbrs.kneighbors_graph(X).toarray()
```

```python
from sklearn import neighbors
from sklearn import preprocessing
# many methods work better on scaled X
X_scaled = preprocessing.scale(X) #
clf1 = neighbors.KNeighborsRegressor(5)#5
train_X = X_scaled[:half]
test_X = X_scaled[half:]
clf1.fit(train_X,train_Y)
```

```python
Y_knn_pred = clf1.predict(test_X)
mse = mean_squared_error(test_Y,Y_knn_pred) ; print(mse)

f, ax = plt.subplots()
ax.plot(test_Y, Y_knn_pred - test_Y,'o',alpha=0.4)
ax.set_title("k-NN Residuals - MSE = %.001f" % mse)
ax.set_xlabel("True Median House Price ($1,000)")
ax.set_ylabel("Residual")
ax.hlines(0,min(test_Y),max(test_Y),color="red")
```
